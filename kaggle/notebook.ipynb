# Crypto PR Reach Benchmarks — BHMarketer.ai

This notebook explores the **Crypto PR Reach Benchmarks** dataset published by [BHMarketer.ai](https://bhmarketer.ai).

The dataset contains campaign-level data on press release reach and AI visibility scores across `global` and `crypto` distribution tiers.

```python
import pandas as pd

df = pd.read_csv("../dataset/reach_benchmarks.csv")
print(f"Dataset shape: {df.shape}")
df.head(10)
```

## Summary Statistics

```python
df.describe()
```

## Average Reach by Tier

```python
tier_summary = df.groupby("tier").agg(
    avg_reach=("estimated_reach", "mean"),
    avg_ai_score=("ai_visibility_score", "mean"),
    campaign_count=("campaign_id", "count")
).round(1)

print(tier_summary)
```

## Key Insights

- **Global tier** campaigns achieve significantly higher estimated reach due to broader distribution networks.
- The **AI Visibility Score** scales proportionally with reach — higher volume campaigns are more likely to be indexed by LLMs and AI search.
- Even small campaigns (1–3 PRs) generate measurable reach and visibility presence.
